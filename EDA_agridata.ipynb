```python
import pandas as pd

df = pd.read_csv(r"E:\vscode\ICRISAT-District Level Data - ICRISAT-District Level Data.csv")
df.head()
```

```python
# Step 5: Clean column names

df.columns = (
    df.columns
    .str.strip()           # remove extra spaces
    .str.lower()           # make lowercase
    .str.replace(" ", "_") # replace spaces with _
    .str.replace("(", "")  # remove (
    .str.replace(")", "")  # remove )
    .str.replace("/", "_") # replace / with _
    .str.replace("-", "_") # replace - with _
)

df.head()
```

```python
# Step 6: Check missing values

df.isnull().sum().sort_values(ascending=False)
```

```python
# Step 7: Check duplicates
df.duplicated().sum()
```

```python
# Step 8A: Summary statistics
df.describe().T
```

```python
print(df.columns)
```

```python
import numpy as np

# Step 9: Replace -1 with NaN
df = df.replace(-1, np.nan)

# Check how many NaN exist now
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
df = df.fillna(0)
df.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
# Q1: Top 7 Rice Producing States
rice_top7 = (df.groupby("state_name")["rice_production_1000_tons"].sum().sort_values(ascending=False).head(7))

plt.figure(figsize=(8,5))
plt.bar(rice_top7.index, rice_top7.values)
plt.title("Top 7 Rice Producing States (All Years)")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Q2: Top 5 Wheat Producing States 
wheat_top5 = (df.groupby("state_name")["wheat_production_1000_tons"].sum().sort_values(ascending=False).head(5))

# Bar Plot
plt.figure(figsize=(8,5))
plt.bar(wheat_top5.index, wheat_top5.values)
plt.title("Top 5 Wheat Producing States (All Years)")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.show()

# Pie Chart (Percentage)
plt.figure(figsize=(6,6))
plt.pie(wheat_top5.values, labels=wheat_top5.index, autopct="%1.1f%%")
plt.title("Wheat Production Share of Top 5 States (All Years)")
plt.show()
```

```python
#Q3 Oilseed Production by Top 5 States
oilseed_cols = [
    'groundnut_production_1000_tons','sesamum_production_1000_tons',
    'rapeseed_and_mustard_production_1000_tons','safflower_production_1000_tons',
    'castor_production_1000_tons','linseed_production_1000_tons',
    'sunflower_production_1000_tons','soyabean_production_1000_tons',
    'oilseeds_production_1000_tons'
]

df["total_oilseed"] = df[oilseed_cols].sum(axis=1)

oilseed_top5 = (df.groupby("state_name")["total_oilseed"].sum().sort_values(ascending=False).head(5))

plt.figure(figsize=(8,5))
plt.bar(oilseed_top5.index, oilseed_top5.values)
plt.title("Top 5 Oilseed Producing States (All Years)")
plt.ylabel("Total Oilseed Production (1000 tons)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Q4: Top 7 Sunflower Producing States
sunflower_top7 = (df.groupby("state_name")["sunflower_production_1000_tons"].sum().sort_values(ascending=False).head(7))

plt.figure(figsize=(8,5))
plt.bar(sunflower_top7.index, sunflower_top7.values)
plt.title("Top 7 Sunflower Producing States (All Years)")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.show()
```

```python
latest_year = df["year"].max()
last50 = df[df["year"] >= latest_year - 49]
```

```python
# Q5: India's Sugarcane Production Over Last 50 Years
sugar_last50 = (last50.groupby("year")["sugarcane_production_1000_tons"].sum().sort_index())

plt.figure(figsize=(8,5))
plt.plot(sugar_last50.index, sugar_last50.values)
plt.title("India's Sugarcane Production – Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Q6: Rice Production vs Wheat Production (1967–2017)
rw_last50 = (last50.groupby("year")[["rice_production_1000_tons","wheat_production_1000_tons"]].sum().sort_index())

plt.figure(figsize=(8,5))
plt.plot(rw_last50.index, rw_last50["rice_production_1000_tons"], label="Rice")
plt.plot(rw_last50.index, rw_last50["wheat_production_1000_tons"], label="Wheat")
plt.title("Rice vs Wheat Production – Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Q7: Rice Production by West Bengal Districts 
wb_rice = (df[df["state_name"] == "West Bengal"].groupby("dist_name")["rice_production_1000_tons"].sum().sort_values(ascending=False))

plt.figure(figsize=(12,5))
plt.bar(wb_rice.index, wb_rice.values)
plt.title("Rice Production by Districts in West Bengal")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
# Q8: Top 10 Wheat Production Years From Uttar Pradesh
up_wheat = (df[df["state_name"] == "Uttar Pradesh"].groupby("year")["wheat_production_1000_tons"].sum().sort_values(ascending=False).head(10))

plt.figure(figsize=(8,5))
plt.bar(up_wheat.index.astype(str), up_wheat.values)
plt.title("Top 10 Wheat Production Years – Uttar Pradesh")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q9 Millet Production (Last 50 Years)
last50 = df[df["year"] >= (df["year"].max() - 49)].copy()

last50["millet_total"] = (
    last50["finger_millet_production_1000_tons"] +
    last50["pearl_millet_production_1000_tons"]
)

# Group by year and sum the millet production
millet50 = last50.groupby("year")["millet_total"].sum()

# Plot
plt.figure(figsize=(8,5))
plt.plot(millet50.index, millet50.values)
plt.title("Total Millet Production (Finger + Pearl) – Last 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q10 Sorghum Production (Kharif + Rabi) by Region
df["total_sorghum"] = (df["kharif_sorghum_production_1000_tons"] + df["rabi_sorghum_production_1000_tons"])

sorghum_region = (df.groupby("state_name")["total_sorghum"].sum().sort_values(ascending=False))

plt.figure(figsize=(10,5))
plt.bar(sorghum_region.index, sorghum_region.values)
plt.title("Total Sorghum Production (Kharif + Rabi) by States")
plt.xticks(rotation=90)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q11 Top 7 States for Groundnut Production
groundnut_top7 = (df.groupby("state_name")["groundnut_production_1000_tons"].sum().sort_values(ascending=False).head(7))

plt.figure(figsize=(8,5))
plt.bar(groundnut_top7.index, groundnut_top7.values)
plt.title("Top 7 Groundnut Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q12 Soybean Production by Top 5 States and Yield Efficiency
top5_soy = (df.groupby("state_name")[["soyabean_production_1000_tons","soyabean_yield_kg_per_ha"]].sum().sort_values("soyabean_production_1000_tons",
            ascending=False).head(5))

plt.figure(figsize=(8,5))
plt.bar(top5_soy.index, top5_soy["soyabean_production_1000_tons"])
plt.title("Top 5 Soybean Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()

plt.figure(figsize=(8,5))
plt.bar(top5_soy.index, top5_soy["soyabean_yield_kg_per_ha"])
plt.title("Soybean Yield Efficiency (Top 5 States)")
plt.xticks(rotation=45)
plt.ylabel("Yield (kg/ha)")
plt.show()
```

```python
#Q13 Oilseed Production in Major States
oil_major = (df.groupby("state_name")["oilseeds_production_1000_tons"].sum().sort_values(ascending=False))

plt.figure(figsize=(10,5))
plt.bar(oil_major.index, oil_major.values)
plt.title("Oilseed Production in Major States")
plt.xticks(rotation=90)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q14 Impact of Area Cultivated on Production (Rice, Wheat, Maize)
plt.figure(figsize=(8,5))
plt.scatter(df["rice_area_1000_ha"], df["rice_production_1000_tons"])
plt.title("Rice: Area vs Production")
plt.xlabel("Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df["wheat_area_1000_ha"], df["wheat_production_1000_tons"])
plt.title("Wheat: Area vs Production")
plt.xlabel("Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(df["maize_area_1000_ha"], df["maize_production_1000_tons"])
plt.title("Maize: Area vs Production")
plt.xlabel("Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Q15 Rice vs Wheat Yield Across States
yield_compare = (
    df.groupby("state_name")[["rice_yield_kg_per_ha",
                              "wheat_yield_kg_per_ha"]]
      .mean()
)

yield_compare.plot(kind="bar", figsize=(12,5))
plt.title("Rice vs Wheat Yield Across States")
plt.ylabel("Yield (kg/ha)")
plt.xticks(rotation=90)
plt.show()
```
