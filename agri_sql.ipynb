```python
#1. Year-wise Trend of Rice Production Across States (Top 3)
query = """
SELECT year, state_name, SUM(rice_production_1000_tons) AS rice_prod
FROM agri_data
GROUP BY year, state_name
ORDER BY rice_prod DESC
LIMIT 3;
"""

pd.read_sql(query, engine)
```

```python
#2. Top 5 Districts by Wheat Yield Increase (Last 5 Years — Simple Version)
query = """
SELECT dist_name, AVG(wheat_yield_kg_per_ha) AS avg_wheat_yield
FROM agri_data
GROUP BY dist_name
ORDER BY avg_wheat_yield DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
#3. States With Highest Oilseed Production
query = """
SELECT state_name, SUM(oilseeds_production_1000_tons) AS total_oilseed
FROM agri_data
GROUP BY state_name
ORDER BY total_oilseed DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
#4. District-wise Area vs Production (Rice, Wheat, Maize)
query = """
SELECT dist_name,
       AVG(rice_area_1000_ha) AS rice_area,
       AVG(rice_production_1000_tons) AS rice_production,
       AVG(wheat_area_1000_ha) AS wheat_area,
       AVG(wheat_production_1000_tons) AS wheat_production,
       AVG(maize_area_1000_ha) AS maize_area,
       AVG(maize_production_1000_tons) AS maize_production
FROM agri_data
GROUP BY dist_name;
"""

pd.read_sql(query, engine)
```

```python
#5. Cotton Production — Top 5 States
query = """
SELECT state_name, SUM(cotton_production_1000_tons) AS cotton_prod
FROM agri_data
GROUP BY state_name
ORDER BY cotton_prod DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
#6. Highest Groundnut Production Districts in 2020
query = """
SELECT dist_name, groundnut_production_1000_tons
FROM agri_data
WHERE groundnut_production_1000_tons IS NOT NULL
ORDER BY groundnut_production_1000_tons DESC
LIMIT 10;
"""
pd.read_sql(query, engine)
```

```python
#7. Annual Average Maize Yield Across States
query = """
SELECT year, AVG(maize_yield_kg_per_ha) AS avg_maize_yield
FROM agri_data
GROUP BY year
ORDER BY year;
"""

pd.read_sql(query, engine)
```

```python
#8. Total Oilseed Area for Each State
query = """
SELECT state_name, SUM(oilseeds_area_1000_ha) AS total_area
FROM agri_data
GROUP BY state_name
ORDER BY total_area DESC;
"""

pd.read_sql(query, engine)
```

```python
#9. Highest Rice Yield Districts
query = """
SELECT dist_name, AVG(rice_yield_kg_per_ha) AS avg_rice_yield
FROM agri_data
GROUP BY dist_name
ORDER BY avg_rice_yield DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

```python
#10. Compare Rice vs Wheat for Top 5 States
query = """
SELECT state_name,
       SUM(rice_production_1000_tons) AS rice_total,
       SUM(wheat_production_1000_tons) AS wheat_total
FROM agri_data
GROUP BY state_name
ORDER BY (rice_total + wheat_total) DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```
